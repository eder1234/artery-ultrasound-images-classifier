```python
#@title # Interface
#@markdown ## Instructions:
#@markdown 1. Press the execution button at the left to start running the code.
#@markdown 2. Upload the images to be classified by the model.
from google.colab import drive
from google.colab import files
import joblib
import cv2
import numpy as np
import matplotlib.pyplot as plt

#drive.mount('/content/drive')

!git clone https://github.com/eder1234/artery-ultrasound-images-classifier.git

#model_filename = '/content/drive/MyDrive/KNN_model.pkl'
model_filename = "/content/artery-ultrasound-images-classifier/KNN_model.pkl"
clf = joblib.load(model_filename)
#scaler_filename = '/content/drive/MyDrive/svd_scaler.pkl'
scaler_filename = "/content/artery-ultrasound-images-classifier/svd_scaler.pkl"
scaler = joblib.load(scaler_filename)

def compute_histogram(image_name):
  # Load the image
  img = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)  # Load as grayscale
  img = cv2.resize(img, (256, 256), interpolation=cv2.INTER_LINEAR)

  if img is None:
    raise ValueError(f"Failed to load image: {image_name}")
  # Compute the histogram
  hist = cv2.calcHist([img], [0], None, [256], [0, 256])

  return hist
  # prompt: Ask the user to drag and drop an image to be classified by the model

uploaded = files.upload()
X = []
for fn in uploaded.keys():
  # predicting images
  path = '/content/' + fn
  histogram = compute_histogram(path)
  histogram = scaler.transform(histogram.reshape(1, -1))
  X.append(histogram.flatten())
  result = clf.predict(np.array(X))
  print(result)
```
